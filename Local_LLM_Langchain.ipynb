## **Local LLM using LangChain Framwork**

To deploy LLMS using open-source models like OpenAI, LLama, Groq etc, it becomes necessary to have access to API keys which often comes with pricing.
Hence, here I am using a model without any need for an API key access to build a local LLM employing LangChain framework.

```python
import transformers
import torch
model = 'EleutherAI/gpt-neo-125m'
tokenizer=transformers.AutoTokenizer.from_pretrained(model)
```

```python
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    max_length=200,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    pad_token_id=tokenizer.eos_token_id,
)

```

```python
from langchain import HuggingFacePipeline,PromptTemplate,LLMChain
llm=HuggingFacePipeline(pipeline=pipeline)
from langchain import PromptTemplate

prompt=PromptTemplate(
    input_variables=["content"],
    template="Tell me a story about {content}."
)

chain=LLMChain(
    llm=llm,
    prompt=prompt
)

print(chain.run('aliens'))
```

If you note the result, the story is very bizzare and gibberish. As the model, we have used is trained on a very small amount of data, it results in such responses.

```python
prompt=PromptTemplate(
    input_variables=["content"],
    template="Tell me a poem about {content}."
)

chain=LLMChain(
    llm=llm,
    prompt=prompt
)

print(chain.run('love'))
```
